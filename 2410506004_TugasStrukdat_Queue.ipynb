```python
#lat1
class QueueList:
    def __init__(self, capacity=10):
        self.queue = []
        self.capacity = capacity

    def enqueue(self, ticket_id):
        if "7" in ticket_id:
            print(f"VIP Customer {ticket_id} langsung mendapatkan kursi tanpa mengantre!")
        else:
            if len(self.queue) < self.capacity:
                self.queue.append(ticket_id)
                print(f"Pelanggan {ticket_id} ditambahkan ke antrian.")
            else:
                print(f"Antrian penuh! Pelanggan {ticket_id} harus menunggu.")

    def dequeue(self):
        if not self.is_empty():
            served = self.queue.pop(0)
            print(f"Pelanggan {served} telah dilayani.")
            return served
        else:
            print("Antrian kosong! Tidak ada pelanggan untuk dilayani.")
            return None

    def front(self):
        return self.queue[0] if not self.is_empty() else None

    def rear(self):
        return self.queue[-1] if not self.is_empty() else None

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

queue = QueueList()

customers = ["AP04", "RA04", "DI04", "AN04", "FA04", "VI04", "JO04", "ME04", "SI04", "LU04", "MA74"]

for customer in customers:
    queue.enqueue(customer)

print("\nAntrian saat ini:", queue.queue)
print("Pelanggan pertama:", queue.front())
print("Pelanggan terakhir:", queue.rear())
print("Jumlah pelanggan dalam antrian:", queue.size())

queue.dequeue()
queue.dequeue()

print("\nAntrian setelah beberapa pelanggan dilayani:", queue.queue)
print("Pelanggan pertama:", queue.front())
print("Pelanggan terakhir:", queue.rear())
print("Jumlah pelanggan dalam antrian:", queue.size())
```

```python
#lat2
class QueueList:
    def __init__(self):
        self.queue = []
        self.fast_track = []

    def enqueue(self, order_id, num_pages):
        if num_pages < 10:
            self.fast_track.append((order_id, num_pages))
            print(f"Pesanan {order_id} ({num_pages} halaman) langsung diproses tanpa mengantre! (Fast Track)")
        else:
            self.queue.append((order_id, num_pages))
            print(f"Pesanan {order_id} ({num_pages} halaman) ditambahkan ke antrian.")

    def dequeue(self):
        if self.fast_track:
            processed = self.fast_track.pop(0)
            print(f"Pesanan {processed[0]} ({processed[1]} halaman) selesai diproses (Fast Track).")
            return processed
        elif self.queue:
            order_id, num_pages = self.queue.pop(0)
            if num_pages > 100:
                print(f"Pesanan {order_id} ({num_pages} halaman) diproses dalam dua tahap.")
                remaining_pages = num_pages - 50
                if remaining_pages > 0:
                    self.queue.insert(0, (order_id, remaining_pages))
            else:
                print(f"Pesanan {order_id} ({num_pages} halaman) selesai diproses.")
            return order_id, num_pages
        else:
            print("Antrian kosong! Tidak ada pesanan yang diproses.")
            return None

    def front(self):
        if self.fast_track:
            return self.fast_track[0]
        return self.queue[0] if self.queue else None

    def rear(self):
        if self.queue:
            return self.queue[-1]
        return self.fast_track[-1] if self.fast_track else None

    def is_empty(self):
        return not self.queue and not self.fast_track

    def size(self):
        return len(self.queue) + len(self.fast_track)


queue = QueueList()

orders = [
    ("APF04", 5),
    ("RAH04", 20),
    ("DIA04", 120),
    ("AND04", 15),
    ("FAT04", 8),
    ("VIC04", 50),
    ("JOH04", 200),
    ("MEG04", 7),
    ("SIT04", 110),
    ("LUK04", 9),
    ("MAR04", 30)
]

for order_id, num_pages in orders:
    queue.enqueue(order_id, num_pages)

print("\nAntrian Fast Track:", queue.fast_track)
print("Antrian Fotokopi:", queue.queue)
print("Pesanan pertama:", queue.front())
print("Pesanan terakhir:", queue.rear())
print("Jumlah pesanan dalam antrian:", queue.size())

queue.dequeue()
queue.dequeue()
queue.dequeue()

print("\nAntrian setelah beberapa pesanan diproses:")
print("Antrian Fast Track:", queue.fast_track)
print("Antrian Fotokopi:", queue.queue)
print("Pesanan pertama:", queue.front())
print("Pesanan terakhir:", queue.rear())
print("Jumlah pesanan dalam antrian:", queue.size())
```

```python
#lat3
class ParkingQueue:
    def __init__(self, max_capacity=15):
        self.queue = []
        self.priority_queue = []
        self.max_capacity = max_capacity

    def enqueue(self, vehicle_number, is_electric=False):
        if is_electric:
            self.priority_queue.append(vehicle_number)
            print(f"Kendaraan {vehicle_number} (Listrik) mendapat Parkir Prioritas!")
        elif len(self.queue) + len(self.priority_queue) < self.max_capacity:
            self.queue.append(vehicle_number)
            print(f"Kendaraan {vehicle_number} masuk parkiran.")
        else:
            print(f"Parkiran penuh! Kendaraan {vehicle_number} harus menunggu.")

    def dequeue(self):
        if self.priority_queue:
            removed = self.priority_queue.pop(0)
            print(f"Kendaraan {removed} (Listrik) keluar dari parkiran.")
            return removed
        elif self.queue:
            removed = self.queue.pop(0)
            print(f"Kendaraan {removed} keluar dari parkiran.")
            return removed
        else:
            print("Parkiran kosong! Tidak ada kendaraan yang keluar.")
            return None

    def front(self):
        if self.priority_queue:
            return self.priority_queue[0]
        return self.queue[0] if self.queue else None

    def rear(self):
        if self.queue:
            return self.queue[-1]
        return self.priority_queue[-1] if self.priority_queue else None

    def is_empty(self):
        return not self.queue and not self.priority_queue

    def size(self):
        return len(self.queue) + len(self.priority_queue)


parking_lot = ParkingQueue()

vehicles = [
    ("MA-04", False),
    ("JA-04", True),
    ("LA-04", False),
    ("BO-04", False),
    ("BA-04", False),
    ("SU-04", False),
    ("SE-04", True),
    ("ME-04", False),
    ("MA-04", False),
    ("LA-04", False),
    ("BO-04", False),
    ("TE-04", False),
    ("DE-04", False),
    ("PA-04", False),
    ("KU-04", False),
    ("CI-04", False),
    ("BE-04", False),
    ("PE-04", False),
]

for vehicle_number, is_electric in vehicles:
    parking_lot.enqueue(vehicle_number, is_electric)

print("\nAntrian Parkir Prioritas:", parking_lot.priority_queue)
print("Antrian Parkir Biasa:", parking_lot.queue)
print("Kendaraan pertama:", parking_lot.front())
print("Kendaraan terakhir:", parking_lot.rear())
print("Jumlah kendaraan dalam parkiran:", parking_lot.size())

parking_lot.dequeue()
parking_lot.dequeue()
parking_lot.dequeue()

print("\nAntrian setelah beberapa kendaraan keluar:")
print("Antrian Parkir Prioritas:", parking_lot.priority_queue)
print("Antrian Parkir Biasa:", parking_lot.queue)
print("Kendaraan pertama:", parking_lot.front())
print("Kendaraan terakhir:", parking_lot.rear())
print("Jumlah kendaraan dalam parkiran:", parking_lot.size())
```

```python
#lat4
class CertificateQueue:
    def __init__(self, max_capacity=20):
        self.regular_queue = []
        self.achievement_queue = []
        self.max_capacity = max_capacity

    def enqueue(self, student_id, is_achiever=False, is_competitor=False):
        if is_competitor:
            print(f"Mahasiswa {student_id} langsung mengambil sertifikat tanpa antri!")
        elif is_achiever:
            self.achievement_queue.append(student_id)
            print(f"Mahasiswa {student_id} (Berprestasi) masuk ke jalur berprestasi!")
        elif len(self.regular_queue) + len(self.achievement_queue) < self.max_capacity:
            self.regular_queue.append(student_id)
            print(f"Mahasiswa {student_id} masuk ke jalur reguler.")
        else:
            print(f"Antrian penuh! Mahasiswa {student_id} harus menunggu.")

    def dequeue(self):
        if self.achievement_queue:
            removed = self.achievement_queue.pop(0)
            print(f"Mahasiswa {removed} (Berprestasi) mengambil sertifikat.")
            return removed
        elif self.regular_queue:
            removed = self.regular_queue.pop(0)
            print(f"Mahasiswa {removed} mengambil sertifikat.")
            return removed
        else:
            print("Antrian kosong! Tidak ada mahasiswa yang mengambil sertifikat.")
            return None

    def front(self):
        if self.achievement_queue:
            return self.achievement_queue[0]
        return self.regular_queue[0] if self.regular_queue else None

    def rear(self):
        if self.regular_queue:
            return self.regular_queue[-1]
        return self.achievement_queue[-1] if self.achievement_queue else None

    def is_empty(self):
        return not self.regular_queue and not self.achievement_queue

    def size(self):
        return len(self.regular_queue) + len(self.achievement_queue)

certificate_queue = CertificateQueue()

students = [
    ("T04", False, False),
    ("S04", True, False),
    ("U04", False, True),
    ("V04", False, False),
    ("W04", False, False),
    ("X04", True, False),
    ("Y04", False, False),
    ("Z04", False, False),
    ("A04", False, False),
    ("B04", False, False),
    ("C04", False, False),
    ("D04", False, False),
    ("E04", False, False),
    ("F04", False, False),
    ("G04", False, False),
    ("H04", False, False),
    ("I04", False, False),
    ("J04", False, False),
    ("K04", False, False),
    ("L04", False, False),
    ("M04", False, False),
]

for student_id, is_achiever, is_competitor in students:
    certificate_queue.enqueue(student_id, is_achiever, is_competitor)

print("\nJalur Mahasiswa Berprestasi:", certificate_queue.achievement_queue)
print("Jalur Mahasiswa Reguler:", certificate_queue.regular_queue)
print("Mahasiswa pertama:", certificate_queue.front())
print("Mahasiswa terakhir:", certificate_queue.rear())
print("Jumlah mahasiswa dalam antrian:", certificate_queue.size())

certificate_queue.dequeue()
certificate_queue.dequeue()
certificate_queue.dequeue()

print("\nAntrian setelah beberapa mahasiswa mengambil sertifikat:")
print("Jalur Mahasiswa Berprestasi:", certificate_queue.achievement_queue)
print("Jalur Mahasiswa Reguler:", certificate_queue.regular_queue)
print("Mahasiswa pertama:", certificate_queue.front())
print("Mahasiswa terakhir:", certificate_queue.rear())
print("Jumlah mahasiswa dalam antrian:", certificate_queue.size())
```
