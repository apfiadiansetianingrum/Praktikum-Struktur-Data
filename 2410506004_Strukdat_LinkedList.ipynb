```python
class CircularLinkedList:
  def __init__(self):
    self.head = None

  def insert_at_end(self, data):
    new_node = Node(data)
    if not self.head:
      self.head = new_node
      new_node.next = self.head
      return
    temp = self.head
    while temp.next != self.head:
      temp = temp.next
    temp.next = new_node
    new_node.next = self.head

  def display(self):
    if not self.head:
      return
    temp = self.head
    while True:
      print(temp.data, end=" -> ")
      temp = temp.next
      if temp == self.head:
        break
    print('(back to head)')
```

```python
#contoh1
#definisi node dalam linked list
class TaskNode:
  def __init__(self, task_name):
    self.task_name = task_name #menyimpan nama tugas
    self.next = None #pointer ketugas berikutnya

#definisi linked list untuk daftar tugas
class TaskList:
  def __init__(self):
    self.head = None #awalnya tidak ada tugas(daftar kosong)

  #menambahkan tugas ke linked list
  def add_task(self, task_name):
    new_task = TaskNode(task_name) #membuat node baru untuk tugas baru
    if not self.head:
      self.head = new_task #jika list kosong, tugas pertama menjadi head
    else:
      current = self.head
      while current.next: #mencari node terakhir dalam list
        current = current.next
      current.next = new_task #menambahkan tugas di akhir list

  #menampilkan daftar tugas
  def show_tasks(self):
    if not self.head:
      print("\n tidak ada tugas hari ini")
      return

    print("\n daftar tugas hari ini:")
    current = self.head
    count = 1
    while current:
      print(f"{count}. {current.task_name}")
      current = current.next
      count += 1

#===program utama(main)===#

#membuat objek daftar tugas
task_list = TaskList()

#meminta pengguna memasukkan jumlah tugas
num_tasks = int(input("masukkan jumlah tugas hari ini: "))

#loop untuk memasukkan tugas kedalam linked list
for i in range(num_tasks + 1):
  task_name = input(f"masukkan tugas ke-{i}: ")
  task_list.add_task(task_name)

#menampilkan daftar tugas
task_list.show_tasks()
```

```python
#contoh 2
#definisi node dalam doubly linked list
class PageNode:
  def __init__(self, url):
    self.url = url #url halaman web
    self.next = None #pointer ke halaman berikutnya
    self.prev = None #pointer ke halaman sebelumnya

#definisi doubly linked list untuk riwayat peramban
class BrowserHistory:
  def __init__(self):
    self.current = None #pointer ke halaman yang sedang dibuka

  #menambahkan halaman ke riwayat
  def visit_page(self, url):
    new_page = PageNode(url)
    if not self.current:
      self.current = new_page #jika riwayat kosong, halaman ini menjadi halaman pertama
    else:
      new_page.prev = self.current
      self.current.next = new_page
      self.current = new_page #pindahkan posisi kehalaman baru

  #menampilkan riwayat kunjungan
  def show_history(self):
    if not self.current:
      print("\n riwayat kosong")
      return

    print("\n riwayat peramban web:")
    temp = self.current
    while temp.prev: #kembali ke halaman pertama
      temp = temp.prev

    count = 1
    while temp:
      print(f"{count}. {temp.url}")
      temp = temp.next
      count += 1

  #navigasi mundur kehalaman sebelumnya
  def go_back(self):
    if self.current and self.current.prev:
      self.current = self.current.prev
      print(f"\nAnda kembali ke {self.current.url}")
    else:
      print("\nTidak ada halaman sebelumnya")

  #navigasi maju ke halaman berikutnya
  def go_forward(self):
    if self.current and self.current.next:
      self.current = self.current.next
      print(f"\nAnda maju ke {self.current.url}")
    else:
      print("\nTidak ada halaman berikutnya")

#===program utama(main)===#
#membuat objek riwayat peramban
history = BrowserHistory()

#meminta pengguna memasukkan jumlah halaman yang dikunjungi
num_pages = int(input("masukkan jumlah halaman yang dikunjungi: "))

#loop untuk memasukkan url halaman kedalam doubly linked list
for i in range(1, num_pages + 1):
  url = input(f"masukkan url halaman ke-{i}: ")
  history.visit_page(url)

#menampilkan riwayat kunjungan
history.show_history()

#menu navigasi pengguna
while True:
  command = input("\ntekan 'b' untuk kembali, tekan 'f' untuk maju, tekan 'x' untuk keluar.\nMasukkan perintah: ").lower()
  if command == "b":
    history.go_back()
  elif command == "f":
    history.go_forward()
  elif command == "x":
    print("\nkeluar dari riwayat peramban")
    break
  else:
    print("\n perintah tidak valid")
```

```python
#contoh 3
# definisi node dalam circular linked list
class SongNode:
  def __init__(self,title):
    self.title = title # Menyimpan judul lagu
    self.next = None # Pointer ke lagu berikutnya

# Definisi Circular Linked List untuk Playlist Lagu
class PlayList:
  def __init__(self):
    self.current = None # Lagu yang sedang diputar

  # Menambahkan lagu ke dalam playlist
  def add_song(self, title):
    new_song = SongNode(title)
    if not self.current:
      self.current = new_song
      self.current.next = self.current # Menghubungkan ke dirinya sendiri (Circular)
    else:
      temp = self.current
      while temp.next != self.current: # Mencari lagu terakhir
        temp = temp.next
      temp.next = new_song # Menambahkan lagu diakhir
      new_song.next = self.current # Menghubungkan kembali ke awal

  # Menampilkan daftar lagu dalam playlist
  def show_playlist(self):
    if not self.current:
      print("\n Playlist kosong!")
      return

    print("\n Playlist Lagu Anda:")
    temp = self.current
    count = 1
    while True:
      print(f"{count}. {temp.title}")
      temp = temp.next
      count += 1
      if temp == self.current: # Berhenti jika sudah kembali ke awal
        break

  # Memuat lagu berikutnya dalam playlist
  def next_song(self):
    if self.current:
      self.current = self.current.next
      print(f"\n Memutar: {self.current.title}")

# ===== Program Utama (Main) ===== #

# Membuat objek playlist
playlist = PlayList()

# Meminta pengguna memasukkan jumlah lagu
num_songs = int(input("\nMasukkan jumlah lagu dalam playlist: "))

# Loop untuk memasukkan lagu ke dalam Circular Linked List
for i in range(1, num_songs + 1):
  title = input(f"Masukkan judul lagu ke-{i}: ")
  playlist.add_song(title)

# Menampilkan daftar lagu
playlist.show_playlist()

# Menu navigasi pengguna
while True:
  command = input("\n Tekan 'n' untuk lagu berikutnya, Tekan 'x' untuk keluar.\nMasukkan perintah: ").lower()

  if command == 'n':
    playlist.next_song()
  elif command == 'x':
    print("\n Keluar dari pemutar musik.")
    break
  else:
    print("\n Perintah tidak dikenal.")
```

```python
#lat3
class Node:
  def __init__(self, data):
    self.data = data # Menyimpan nilai node
    self.next = None # Pointer ke node berikutnya

# Definisi Singly Linked List dengan Berbagai Operaasi
class LinkedList:
  def __init__(self):
    self.head = None # Awalnya, daftar kosong

  # Menambahkan elemen di awal
  def add_at_beginning(self, data):
    new_node = Node(data)
    new_node.next = self.head
    self.head = new_node

  # Menambahkan elemen di akhir
  def add_at_end(self, data):
    new_node = Node(data)
    if not self.head:
      self.head = new_node
      return
    temp = self.head
    while temp.next:
      temp = temp.next
    temp.next = new_node

  # Menambahkan elemen di posisi tertentu
  def add_at_position(self, position, data):
    if position < 0:
      self.add_at_beginning(data)
      return
    temp = self.head
    for _ in range(position - 1):
      if not temp:
        print(" Posisi melebihi panjang daftar!")
        return
      temp = temp.next
    new_node.next = temp.next
    temp.next = new_node

  # Menghapus elemen di awal
  def delete_at_beginning(self):
    if not self.head:
      print(" Daftar kosong, tidak ada yang bisa dihapus!")
      return
    self.head = self.head.next

  # Menghapus elemen di akhir
  def delete_at_end(self):
    if not self.head:
      print(" Daftra kosong, tidak ada yang bisa dihapus!")
      return
    if not self.head.next:
      self.head = None
      return
    temp = self.head
    while temp.next.next:
      temp = temp.next
    temp.next = None

  # Menghapus elemen di posisi tertentu
  def delete_at_position(self, position):
    if not self.head:
      print(" Daftar kosong, tidak ada yang bisa dihapus!")
      return
    if position == 0:
      self.head = self.head.next
      return
    temp = self.head
    for _ in range(position - 1):
      if not temp or not temp.next:
        print("⚠ Posisi tidak ditemukan!")
        return
      temp = temp.next
    temp.next = temp.next.next

  # Mencari elemen dalam Linked List
  def search(self, key):
    temp = self.head
    while temp:
      if temp.data == key:
        print(f" Elemen {key} ditemukan dalam daftar!")
        return
      temp = temp.next
    print(f" Elemen {key} tidak ditemukan dalam daftar.")

  # Membalik (reverse) Linked List
  def reverse(self):
    prev = None
    current = self.head
    while current:
      next_node = current.next
      current.next = prev
      prev = current
      current = next_node
    self.head = prev

  # Menampilkan Linked List
  def display(self):
    if not self.head:
      print(" Daftar kosong!")
      return
    temp = self.head
    while temp:
      print(temp.data, end=" -> ")
      temp = temp.next
    print("None")

# ===== Program Utama (Main) ===== #

# Membuat objek Linked List
ll = LinkedList()

# Meminta pengguna memasukkan elemen awal
num_elements = int(input("\nMasukkan jumlah awal elemen dalam daftar: "))
for i in range(num_elements):
  data = int(input(f"Masukkan data ke-{i+1}: "))
  ll.add_at_end(data)

# Menu Operasi Linked List
while True:
  print("\n Pilih Operasi:")
  print("1. Tambah elemen di awal")
  print("2. Tambah elemen di akhir")
  print("3. Tambah elemen di posisi tertentu")
  print("4. Hapus elemen di awal")
  print("5. Hapus elemen di akhir")
  print("6. Hapus elemen di posisi tertentu")
  print("7. Cari elemen dalam daftar")
  print("8. Balikkan daftar")
  print("9. Tampilkan daftar")
  print("0. Keluar")

  choice = int(input("\nMasukkan pilihan: "))

  if choice == 1:
    data = int(input("Masukkan data yang ingin ditampilkan di awal: "))
    ll.add_at_beginning(data)
  elif choice == 2:
    data = int(input("Masukkan data yang ingin ditampilkan di akhir: "))
    ll.add_at_end(data)
  elif choice == 3:
    position = int(input("Masukkan posisi untuk menambahkan data: "))
    data = int(input("Masukkan data yang ingin ditambahkan: "))
    ll.add_at_position(position, data)
  elif choice == 4:
    ll.delete_at_beginning()
  elif choice == 5:
    ll.delete_at_end()
  elif choice == 6:
    position = int(input("Masukkan posisi elemen yang ingin dihapus: "))
    ll.delete_at_position(position)
  elif choice == 7:
    key = int(input("Masukkan elemen yang ingin dicari: "))
    ll.search(key)
  elif choice == 8:
    ll.reverse()
    print(" Daftar berhasil diperbaiki!")
  elif choice == 9:
    ll.display()
  elif choice == 0:
    print("\n Keluar dari program.")
    break
  else:
    print("\n Pilih tidak valid, silahkan coba lagi!")
```

```python
#lat2

#definisi node dalam Doubly
class FolderNode:
  def __init__(self, folder):
    self.folder = folder
    self.next = None
    self.prev = None

#definisi untuk riwayat peramban
class FolderHistory:
  def __init__(self):
    self.current = None

#menambahkan halaman baru
  def navigasi_folder(self, folder):
    new_folder = FolderNode(folder)
    if not self.current:
      self.current = new_folder #Jika riwayat kosong, halaman ini menjadi halaman pertama
    else:
      new_folder.prev = self.current
      self.current.next = new_folder
      self.current = new_folder #Pindahkan posisi ke halaman baru

  #Menampilkan riwayat kunjungan
  def show_history(self):
    if not self.current:
      print("\n Riwayat Folder Kosong.")
      return

    print("\n Riwayat Navigasi Folder:")
    temp = self.current
    while temp.prev: #Kembali ke halaman pertama
      temp = temp.prev

    count = 1
    while temp:
      print(f"{count}. {temp.folder}")
      temp = temp.next
      count += 1

  #Navigasi mundur ke halaman sebelumnya
  def go_back(self):
    if self.current and self.current.prev:
      self.current = self.current.prev
      print(f"\n Anda kembali ke: {self.current.folder}")
    else:
      print("\n Tidak bisa kembali, ini adalah folder pertama!")

  #Navigasi maju ke halaman berikutnya
  def go_forward(self):
    if self.current and self.current.prev:
      self.current = self.current.next
      print(f"\n Anda maju ke: {self.current.folder}")
    else:
      print(f"\n Tidak bisa maju, ini adalah folder terakhir!")

# ==== Program ==== #

#Membuat objek riwayat peramban
history = FolderHistory()

#Meminta pengguna memasukan jumlah halaman yang dikunjungi
npm_digit_terakhir = int(input("\nMasukkan NPM digit terakhir: "))
num_folders = npm_digit_terakhir

#Loop untuk memasukkan URL halaman ke dalam Doubly Linked List
for i in range(1, num_folders + 1):
  folder = input(f"Masukkan folder halaman ke-{i}: ")
  history.navigasi_folder(folder)

#Menampilkan riwayat yang telah dimasukkan
history.show_history()

#Menu navigasi pengguna
while True:
  command = input("\n Tekan 'b' untuk kembali, Tekan 'f' untuk maju, Tekan 'x' untuk keluar.\nMasukkan perintah: ").lower()

  if command == 'b':
    history.go_back()
  elif command == 'f':
    history.go_forward()
  elif command == 'x':
    print("\n Keluar dari riwayat navigsi folder.")
    break
  else:
    print("\n Perintah tidak dikenal.")
```
